```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'marital_Together']
data_scaled = data.copy()
data_scaled[cols_for_clustering] = scaler.fit_transform(data[cols_for_clustering])
```

```python
data_scaled[cols_for_clustering]
```

```python
data_scaled[cols_for_clustering].head()
```

```python
data_scaled[cols_for_clustering].tail()
```

```python
data_scaled[cols_for_clustering].describe()
```

```python
X = data_scaled[cols_for_clustering]
inertia_list = []
for K in range(2,20):
    inertia = KMeans(n_clusters=K, random_state=8).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[5,5])
plt.plot(range(2,20), inertia_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Inertia / Number of Clusters")
plt.xlabel("Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(data_scaled[cols_for_clustering])
data_scaled['pc1'] = pca_res[:,0]
data_scaled['pc2'] = pca_res[:,1]
```

```python
model = KMeans(n_clusters=5, random_state = 8)
model.fit(data_scaled[cols_for_clustering])
data_scaled['Cluster'] = model.predict(data_scaled[cols_for_clustering])
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='pc1', y='pc2', data=data_scaled, hue='Cluster', palette='ocean')
plt.title('Clustered  Visualization')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.legend(title='Clusters')
```
